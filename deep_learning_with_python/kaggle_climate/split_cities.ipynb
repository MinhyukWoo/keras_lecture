```python
import pandas as pd
import os

dataset = pd.read_csv(os.path.join("dataset", "weather_prediction_dataset.csv"))
dataset
```

```python
for col in dataset.columns:
    print(col)
```

```python
if "MONTH" in dataset.columns:
    dataset = dataset.drop(columns=["MONTH"])
dataset
```

```python
city_dataset_list = []
prev_col_first_name = ""
city_dataset = None
for col in dataset.columns[1:]:
    col_first_name = col.split('_')[0]
    if col_first_name == prev_col_first_name:
        city_dataset = pd.concat([city_dataset, dataset[col]], axis=1)
    else:
        if city_dataset is not None:
            city_dataset_list.append(city_dataset)
        city_dataset = pd.Series(dataset["DATE"])
        prev_col_first_name = col_first_name
city_dataset_list.append(city_dataset)

```

```python
city_dataset_list[-1]
```

```python
for city_dataset in city_dataset_list:
    city_first_name = city_dataset.columns[1].split("_")[0]
    city_dataset.to_csv(os.path.join("dataset", "sorted", city_first_name + ".csv"), index=False)
```
